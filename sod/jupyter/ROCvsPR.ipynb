```python
from sklearn.metrics import roc_curve, average_precision_score, \
    precision_recall_curve, roc_auc_score, auc
```

```python
import numpy as np
%matplotlib inline
print('Unbalanced classes, 10 positives all correctly classified,')
print('10000 negatives of which 1000 NOT correctly classified')
y_true = np.append(np.zeros(10000), np.ones(10))
y_pred = np.append(np.zeros(9000), np.ones(1010))

fpr, tpr, thresholds = roc_curve(y_true, y_pred)

import matplotlib.pyplot as plt
plt.plot(fpr, tpr)
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC Curve: AUC: %f' % roc_auc_score(y_true, y_pred))
plt.show()


p, r, th = precision_recall_curve(y_true, y_pred)
plt.plot(r, p)
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('PR Curve APS: %f' % average_precision_score(y_true, y_pred))
plt.show()
```

```python
import numpy as np
%matplotlib inline
print('Balanced classes, 10000 positives all correctly classified,')
print('10000 negatives of which 1000 NOT correctly classified')
y_true = np.append(np.zeros(10000), np.ones(10000))
y_pred = np.append(np.zeros(9000), np.ones(11000))

#   Classified as:
#   I     O
# I 9,000  1,000
# O     0 10,000

fpr, tpr, thresholds = roc_curve(y_true, y_pred)

import matplotlib.pyplot as plt
plt.plot(fpr, tpr)
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC Curve: AUC: %f' % roc_auc_score(y_true, y_pred))
plt.show()


p, r, th = precision_recall_curve(y_true, y_pred)
plt.plot(r, p)
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('PR Curve APS: %f' % average_precision_score(y_true, y_pred))
plt.show()
```

```python
import numpy as np
%matplotlib inline
print('Balanced classes, 10000 positives 9000 correctly classified,')
print('10000 negatives all correctly classified')
y_true = np.append(np.zeros(10000), np.ones(10000))
y_pred = np.append(np.zeros(11000), np.ones(9000))

# confusion matrix:

#   Classified as:
#   I      O
# I 10,000 0
# O  1,000 9,000

fpr, tpr, thresholds = roc_curve(y_true, y_pred)

import matplotlib.pyplot as plt
plt.plot(fpr, tpr)
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC Curve: AUC: %f' % roc_auc_score(y_true, y_pred))
plt.show()

# assert auc(fpr, tpr) == roc_auc_score(y_true, y_pred)

p, r, th = precision_recall_curve(y_true, y_pred)
plt.plot(r, p)
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('PR Curve APS: %f' % average_precision_score(y_true, y_pred))
plt.show()
```
